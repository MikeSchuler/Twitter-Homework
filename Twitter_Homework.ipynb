```python
import tweepy
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import style
from datetime import datetime
import time
style.use('ggplot')

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key,
                    consumer_secret,
                    access_token,
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
target_user1 = "@NYtimes"
target_user2 = "@BBC"
target_user3 = "@CNN"
target_user4 = "@FoxNews"
target_user5 = "@CBSNews"

sentiments1 = []
sentiments2 = []
sentiments3 = []
sentiments4 = []
sentiments5 = []

counter = 1

oldest_tweet = None

# for user in 
for x in range(5):

    # Get all tweets from home feed
    public_tweets = api.user_timeline(target_user1, max_id = oldest_tweet)
    
    # Loop through all tweets 
    for tweet in public_tweets:
        
        # Run Vader Analysis on each tweet
        results = analyzer.polarity_scores(tweet["text"])
        compound = results["compound"]
        pos = results["pos"]
        neu = results["neu"]
        neg = results["neg"]
        
        # Get Tweet ID, subtract 1, and assign to oldest_tweet
        oldest_tweet = tweet['id'] - 1
        
        # Add sentiments for each tweet into a list
        sentiments1.append({"Date": tweet["created_at"], 
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": counter,
                           "User": tweet['user']['screen_name']})
        
        # Add to counter 
        counter += 1
```

```python
# @nytimes, @BBC, @CNN, @FoxNews, @CBSNews

target_users = ("@NYtimes", "@BBC", "@CNN", "@FoxNews", "@CBSNews")

# List for dictionaries of results
results_list = []

# Loop through each user
for user in target_users:

    # Variables for holding sentiments
    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []

    # Loop through 100 tweets
    for x in range(5):

        # Get all tweets from home feed
        public_tweets = api.user_timeline(user, page=x, count=100)

        # Loop through all tweets
        for tweet in public_tweets:

            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]

            # Add each value to the appropriate list
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)

    # Create a dictionaty of results
    user_results = {
        "Username": user,
        "Compound Score": np.mean(compound_list),
        "Postive Score": np.mean(positive_list),
        "Neutral Score": np.mean(neutral_list),
        "Negative Score": np.mean(negative_list)
    }

    # Append dictionary to list
    results_list.append(user_results)

    # Print the Averages
    print(f"User: {user}")
    print(f"Compound: {np.mean(compound_list):.3f}")
    print(f"Positive: {np.mean(positive_list):.3f}")
    print(f"Neutral: {np.mean(neutral_list):.3f}")
    print(f"Negative: {np.mean(negative_list):.3f}")
```

```python
# Create DataFrame from Results List
results_df = pd.DataFrame(results_list).set_index("Username").round(3)
results_df
```

```python
colors = ['blue', 'yellow', 'red', 'green', 'purple']

date = time.strftime("%m/%d/%Y")

plt.bar(results_df.index, results_df['Compound Score'], color=colors, width=1)
plt.title("Overall Media Sentiment on Twitter (" + date + ")")
plt.ylabel("Tweet Polarity")
plt.grid(False)
plt.show()
```
